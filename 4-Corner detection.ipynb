```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
img = cv2.imread('AR1.png')
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

corners = cv2.goodFeaturesToTrack(gray,6,0.02,50)
corners = np.int0(corners)
for i in corners:
    x,y = i.ravel()
    cv2.circle(img,(x,y),10,255,-1)

plt.imshow(img)
plt.show()
```

```python
hull = cv2.convexHull(corners)
img = cv2.imread('AR1.png')
for i in hull:
    x,y = i.ravel()
    cv2.circle(img,(x,y),10,255,-1)
plt.imshow(img)
plt.show()
```
